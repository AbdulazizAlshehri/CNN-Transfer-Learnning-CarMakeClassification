```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
gen = ImageDataGenerator(rotation_range=20, width_shift_range=0.1, height_shift_range=0.1, shear_range=0.05, zoom_range=0.3, 
                               channel_shift_range=100., horizontal_flip=True)

for i in list(range(101, 197)):
    path = '/content/drive/MyDrive/Colab Notebooks/dataset/train_spl/' + str(i) + '/'
    image_list = []

    error = 0
    print('folder #' + str(i))
    all_images_names = os.listdir(path)
    for image_name in all_images_names: 

        image_path = path + image_name
        image = np.expand_dims(plt.imread(image_path),0)

        try:
            aug_iter = gen.flow(image, save_to_dir=path, save_prefix='aug-' + image_name, save_format='jpg')
            [next(aug_iter)[0].astype(np.uint8) for i in range(5)]
        except:
            error+=1

    if error > 0:
        print(' - errors: ' + str(error))
        
        
```
